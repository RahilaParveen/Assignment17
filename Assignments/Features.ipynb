```python
surveys = pd.read_csv('surveys.csv', index_col='record_id')
```

```python
surveys.head()
```

```python
surveys.weight.fillna(surveys.weight.mean(),inplace=True)
```

```python
surveys.info()
```

```python
surveys.hindfoot_length.fillna(surveys.hindfoot_length.mean(),inplace=True)
```

```python
surveys.dropna(subset=['species_id','sex'], inplace=True)
```

```python
surveys.info()
```

```python
survey_tmp = surveys[['sex']]
```

```python
survey_tmp.head()
```

```python
ohe_sex_df = pd.get_dummies(survey_tmp,prefix=['sex'])
```

```python
ohe_sex_df.head()
```

```python
surveys=pd.concat([surveys,ohe_sex_df],axis=1)
```

```python
surveys.head()
```

```python
surveys.drop(columns=['sex'],inplace=True)
```

```python
surveys.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
```

```python
X = surveys.drop(columns='species_id')
```

```python
X.head()
```

```python
X_scaled = pd.DataFrame(mms.fit_transform(X))
```

```python
X_scaled.head()
```

```python
X_scaled.index = X.index
```

```python
X.head()
```

```python
X_scaled.columns = X.columns
```

```python
X_scaled.head()
```

```python
X_scaled.tail()
```

```python
surveys_scaled=pd.concat([X_scaled,surveys['species_id']],axis=1)
```

```python
surveys_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(surveys_scaled.iloc[:,:-1],\
                                                    surveys_scaled.iloc[:,-1:],\
                                                   test_size=0.2)
```

```python
X_train.info()
```

```python
X_test.info()
```

```python
y_train.info()
```

```python
y_test.info()
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X=X_train,y=y_train)
# model.fit(X=X_train,y=y_train.values.ravel())
```

```python
y_test['species_pred']=model.predict(X_test).ravel()
```

```python
y_test.columns=['species_true','species_pred']
```

```python
y_test.head()
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(f"""accuracy score: {accuracy_score(y_test.species_true.values, 
                                        y_test.species_pred.values)}""")
```
