```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,\
                                                   test_size=0.2,\
                                                   random_state=7)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.losses import categorical_crossentropy
from keras.metrics import categorical_accuracy, mean_absolute_error, mse
from keras.optimizers import Adam
```

```python
import numpy as np
seed = 7
np.random.seed(seed)
```

```python
model = Sequential()
model.add(Dense(1, input_shape=(4,), activation='relu'))
model.add(Dense(3, activation='softmax'))
model.compile(Adam(lr=0.01), loss='categorical_crossentropy', \
              metrics=['categorical_accuracy', 'accuracy'])
from keras.utils import to_categorical
history = model.fit(X_train, to_categorical(y_train), epochs=1000)
```

```python
from sklearn.metrics import accuracy_score
print(f'Accuracy: {accuracy_score(y_test.species_id.values, y_test.pred.values)}')
```
