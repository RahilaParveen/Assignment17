```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300, 543, 1075, 2862, 5979, 10289, 14958]
```

```python
%matplotlib inline
```

```python
plt.plot(years, \
         gdp, \
         color='blue', \
         marker='o', \
         linestyle='solid')

plt.title('nominal gdp')
plt.ylabel('billions of $')
plt.xlabel('year')
```

```python
plt.title('nominal gdp')
```

```python
plt.bar(years, \
         gdp, \
         color='red')
```

```python
plt.figure(figsize=[12,9])
plt.bar(years, \
         gdp, \
         color='red')
```

```python
plt.bar(years,
       gdp,
       width=5.0)
```

```python
plt.bar(years, \
         gdp, \
         color='red',\
       label='gdp',\
       width=5)
plt.legend(loc='best')
```

```python
meanScore=[12, 13, 14, 12, 15, 20, 15]
classSize=[20, 25, 22, 40, 51, 15, 19]
```

```python
plt.xlabel('No of students')
plt.ylabel('Avg test score')

plt.scatter(classSize,
            meanScore,
           marker='o',
           color='green')
```

```python
plt.xlabel('No of students')
plt.ylabel('Avg test score')
meanScore=[12, 13, 14, 12, 15, 20, 15]
classSize=[20, 25, 22, 40, 51, 15, 19]
s=[10*2**n for n in range(len(meanScore))]
plt.scatter(classSize,meanScore,marker='o', c = 'green', s=s)
plt.show()
```
